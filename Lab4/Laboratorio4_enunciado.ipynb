**Distplot:**

````python
def distplot_grades(dataframe):
  
  data = [dataframe["math score"].tolist(),
          dataframe["reading score"].tolist(),
          dataframe["writing score"].tolist()]
  labels = ['math score', 'reading score', 'writing score']
  colors = ['blue', 'red', 'green']
  fig = ff.create_distplot(data, labels, colors=colors,show_hist=False, show_rug=False)
  fig.show()

````

````python
def histogram_boxplot_grades(dataframe):
  df_melted = pd.melt(dataframe[['math score', 'reading score', 'writing score']], var_name='subject', value_name='score')
  fig = px.histogram(df_melted,x='score',color='subject', marginal='box', nbins=14, barmode='group')
  fig.show()
````

### 1.3.4 Tabla de Resumen [1 punto]

Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificaci칩n a trav칠s de un mapeo (investigar el m칠todo `map()`) de la variable `parental level of education`  seg칰n la siguiente conversi칩n: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [ ] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como gu칤a para desarrollar este punto.

````python
def alumnos_promedio(dataframe_in):
  """
  Add a GPA column (mean of scores rounded to 2 decimal places) to dataframe_in
  ------------------------------------------------------------------
  Inputs:
  dataframe_in (pd.dataframe): the dataframe with the scores.
  ------------------------------------------------------------------
  Outputs:
    dataframe_in (pd.dataframe): the same dataframe as Input with GPA column.
  """

  proms = np.mean(dataframe_in[["math score", 'reading score', 'writing score']].to_numpy(), axis=1)
  dataframe_in['GPA'] = proms.round(decimals = 2)

  return dataframe_in
````

````python
def PLOE_mapping(dataframe):
  data = dataframe['parental level of education']
  data_mapped = data.map({'some high school': 'school', 
                          'some college': 'school',
                          'high school': 'school',
                          "bachelor's degree" : 'college',
                          "associate's degree" : "college",
                          "master's degree" : "postgraduate"})
  dataframe['simple parental level of education'] = data_mapped
  return dataframe
  
````

````python
df_grades = alumnos_promedio(df_grades)
df_grades.head()
````

````python
df_grades = PLOE_mapping(df_grades)
df_grades.head()
````

````python
def group_by_grades(dataframe):
  
  # Agrupar por 'race/ethnicity', 'simple parental level of education' calculando promedios y conteo de nombres. Se renombra la columna de conteo de nombres a 'count'
  grouped = dataframe.groupby(by=['race/ethnicity', 'simple parental level of education']) \
                      .agg({'math score': 'mean', 'reading score': 'mean', 'writing score': 'mean', 'GPA': 'mean', 'names':'count'})\
                      .rename(columns= {'names': 'count'})\
                      .reset_index()

  # Calcular porcentaje a 2 decimales
  grouped['percentage'] = grouped[['count']].apply(lambda x: (100*x/x.sum()).round(2)).astype(str) + ' %' 

  # Redondear N칰mericos
  grouped = grouped.round(2)
  return grouped
````

````python
group_by_grades(df_grades)
````
